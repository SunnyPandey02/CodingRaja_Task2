# Collaborative filtering based recommender system

```python
movies['title'] = movies['title'].str.replace(r'\s*\(\d{4}\)', '', regex=True)
movies.head()
```

```python
x=movies.groupby('userId').count()['rating']>500
literate_users=x[x].index
```

```python
filtered_rating=movies[movies['userId'].isin(literate_users)]
```

```python
y=filtered_rating.groupby('title').count()['rating']>=250
famous_movies=y[y].index
```

```python

final_ratings=filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
final_ratings.drop_duplicates().shape
```

```python
pt=final_ratings.pivot_table(index='title',columns='userId',values='rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
# 4753 famous books and 7441 are the literate users
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(movie_name):
    # index fetch
    index = np.where(pt.index==movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = movies[movies['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
#         item.extend(list(temp_df.drop_duplicates('title')['num_rating'].values))
#         item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("xXx: State of the Union")
```

```python
recommend("(500) Days of Summer")
```

```python
recommend("Toy Story")
```

```python
recommend("Blade: Trinity")
```
